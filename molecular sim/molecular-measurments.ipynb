```python
'''f(r) and B2 of He'''
import matplotlib.pyplot as plt
def He_U(r_U):
    return 4*epsilon[0]*((sigma[0]/r_U)**(12)-(sigma[0]/r_U)**(6))
sigma=[2.56,3.75,3.4,4.07]
epsilon=[1.41*10**-22,1.32*10**-21,1.66*10**-21,3.04*10**-21]
r_U=np.linspace(0.1,20,100)


def He_f(r_f):
    return (1-np.exp(-He_U(r_f)/(k*T)))*r_f**2
r_f=np.linspace(0.1,20,100)
T=600
k=1.3*10**-23

plt.plot(r_f,He_f(r_f))
plt.xlim([0,20])
plt.title('f(r) for He')
plt.xlabel('r in Å')
plt.ylabel('f(r)')
y_points_list=[]
n = 10000 # number of guesses
xlim1 = 0 #x limit
xlim2 = sigma[0]#x 
min_xlim1=sigma[0]
min_xlim2=20
ymax = 3.843636271365093 #max y, min zero
'''ymax and ymin where found using np.min(y_points_list)'''
ymin = 0.4748574441910291 
s = 0 # initialize counter variable
p=0
A = (xlim2 - xlim1)*ymax # area A
A_2 = (min_xlim2 - min_xlim1)*ymin
y_points=He_f(r_f)
y_points_list.append(y_points)
for i in range(n): # for every value of n
    Rx = (xlim2-xlim1)*np.random.rand() + xlim1 # Set Rx, a random number
    Ry = ymax*np.random.rand()# Set Ry, a random number
    minRx=(min_xlim2-min_xlim1)*np.random.rand() + min_xlim1
    minRy=ymin*np.random.rand()
    if Ry <= He_f(Rx): # if we are below the curve
        s = s+1 # increment counter
    if minRy <= -He_f(minRx):
        p = p+1
    av_f = (A*s)/n # ratio below curve * area = integral estimate
    av_f2 = (A_2*p)/n
    total=av_f -av_f2
print(total)
B2_He=total*np.pi
print(B2_He)
```

```python
'''numerical intergration of He'''
def Intergral_calc(r_f):
    return integrate.quad(He_f, 0, 20)
inter=Intergral_calc(r_f)
print(inter)
```
